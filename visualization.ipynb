```python
fig = plt.figure(figsize=(6,4),edgecolor='#9031dd',linewidth=1)   
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x,y)
```

```python
fig = plt.figure(figsize=(6,4))   
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x,y, marker='o', markersize=4, markerfacecolor='blue',
         markeredgecolor='green', markeredgewidth=2)
```

```python
fig = plt.figure(figsize=(6,4))   
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x,y, linestyle=':', color='red', linewidth=1)
```

```python
fig = plt.figure(figsize=(6,4))   
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x,y, linestyle='--', color='blue', linewidth=1, label='sin(x)')
plt.title('time series')
plt.xlabel('time')
plt.ylabel('value')
plt.legend()
```

```python
fig = plt.figure(figsize=(6,2))   
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
plt.plot(x,y, linestyle='--', color='blue', linewidth=1, label='sin(x)')
plt.plot(x,z, linestyle=':', color='red', linewidth=1, label='cos(x)')
plt.title('time series', loc='left', 
          pad=10, fontdict={'fontsize': 16,
                             'color': 'blue',
                             'weight': 'bold'})
plt.xlabel('time')
plt.ylabel('value')
plt.legend()
```

```python
fig = plt.figure(figsize=(6,4))   
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
plt.plot(x,y, linestyle='--', color='blue', linewidth=1, label='sin(x)')
plt.plot(x,z, linestyle=':', color='red', linewidth=1, label='cos(x)')
plt.xlim(-1,12)
plt.ylim(-1.5,1.5)
```

```python
rainfall = [0.8, 0.6, 0, 0,
            0.1, 0.2, 1.2, 1.5, 
            0.1, 0.3, 0.6, 0.8]
months = ['Jan', 'Feb', 'Mar', 'Apr', 
          'May', 'Jun', 'Jul', 'Aug',
          'Sep', 'Oct', 'Nov', 'Dec']
print(len(rainfall), len(months))
fig = plt.figure(figsize=(6,4))
plt.bar(months, rainfall, color='blue', hatch='//')
plt.title('Rainfall in 2019')
plt.xlabel('Month')
plt.ylabel('Rainfall (mm)')
```
